```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
ds = load_digits()
```

```python
print(ds.data)
print(ds.target)
```

```python
print(ds.data.shape)
data_length = len(ds.images)
print(data_length)
```

```python
n= 8#image no.
plt.gray()
plt.matshow(ds.images[n])
plt.show()

ds.images[n]
```

## data seperation

```python
x= ds.images.reshape(data_length,-1)
x
```

```python
y=  ds.target
y
```

## Train  and Test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y, test_size = 0.25, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

## training

```python
from sklearn import svm
model = svm.SVC(gamma=0.001)
model.fit(x_train,y_train)
```

```python
n=5
result = model.predict(ds.images[n].reshape(1,-1))
plt.imshow(ds.images[n],cmap= plt.cm.gray, interpolation= 'nearest')
print(result)
print("\n")

plt.axis('on')
plt.title('%i' %result)
plt.show()
```

```python
y_pred= model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy of the model is : {0}".format(accuracy_score(y_test,y_pred)*100))
```

## play with different method

```python
from sklearn import svm
model1= svm.SVC(kernel='linear')
model2= svm.SVC(kernel='rbf')
model3= svm.SVC(gamma=0.001)
model4= svm.SVC(gamma=.001,C=0.1 )

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)

y_predModel1 = model1.predict(x_test)
y_predModel2 = model2.predict(x_test)
y_predModel3 = model3.predict(x_test)
y_predModel4 = model4.predict(x_test)

print("Accuracy of the Model 1: {0}%".format(accuracy_score(y_test, y_predModel1)*100))
print("Accuracy of the Model 2: {0}%".format(accuracy_score(y_test, y_predModel2)*100))
print("Accuracy of the Model 3: {0}%".format(accuracy_score(y_test, y_predModel3)*100))
print("Accuracy of the Model 4: {0}%".format(accuracy_score(y_test, y_predModel4)*100))
```
