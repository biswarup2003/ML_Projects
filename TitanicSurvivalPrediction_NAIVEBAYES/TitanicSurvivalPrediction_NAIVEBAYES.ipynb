```python
import pandas as pd
import numpy as np
```

```python
ds= pd.read_csv("titanicsurvival.csv")
print(ds)
```

```python
ds.shape
```

```python
ds.head(5)
```

```python
mean_age= ds['Age'].mean()
ds['Age']=ds['Age'].fillna(mean_age)
```

```python
ds.head(5)
```

```python
ds.head(10)
```

```python
# Drop rows where 'Sex' has missing values
ds.dropna(subset=['Sex'], inplace=True)
ds['Sex'] = ds['Sex'].map({'female': 0, 'male': 1}).astype(int)
```

```python
ds.head(5)
```

```python
x= ds.drop('Survived', axis='columns')
x
```

```python
y= ds['Survived']
y
```

```python
y.head(5)
```

```python
x.columns[x.isna().any()]
```

```python
x.Age= x.Age.fillna(x.Age.mean())
```

```python
x.columns[x.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25, random_state= 0)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
x.columns
```

```python
pclass= int(input("Enter the Pclass no. : "))
sex= int(input("Enter the Gender (Female=0 & Male=1) : "))
try:
    if(sex not in [0,1]):
        raise ValueError("Invalid input! Please enter 0 for Female or 1 for Male.")
except ValueError as e:
    print(e)

age=  int(input("Enter the Age : "))
fare= int(input("Enter Fare : "))

new_person=[[pclass,sex,age,fare]]
result= model.predict(new_person)
print(result)

if(result==1):
    print("The Person might be servived ")
else:
    print("The Person might not be servived ")
```

```python
y_pred= model.predict(x_test)
print(np.column_stack((y_pred,y_test)))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm= confusion_matrix(y_test,y_pred)
print(cm)
print("The accurecy of the model is: {0}".format(accuracy_score(y_test,y_pred)*100))
```
