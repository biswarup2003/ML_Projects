```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
ds = load_iris()
```

```python
x= pd.DataFrame(ds.data, columns=ds.feature_names)
x
```

```python
y= ds.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size= 0.25, random_state = 0)
print(x_train.shape)
print(x_test.shape)
```

## best max_depth Value

```python
accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1,10):
    model = DecisionTreeClassifier(max_depth=i, random_state = 0)
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    score= accuracy_score(y_test,pred)
    accuracy.append(score)

plt.plot(range(1,10),accuracy,marker='o')
plt.xlabel('Prediction')
plt.ylabel('Accuracy_score')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(criterion = 'entropy',max_depth = 3, random_state = 0)
model2 = DecisionTreeClassifier(criterion = 'entropy',max_depth = 5, random_state = 0)
model3 = DecisionTreeClassifier(criterion = 'entropy',max_depth = 9, random_state = 0)
model4 = DecisionTreeClassifier(criterion = 'entropy',max_depth = 2, random_state = 0)
model5 = DecisionTreeClassifier(criterion = 'entropy',max_depth = 1, random_state = 0)

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)
model5.fit(x_train,y_train)
```

```python
y_pred_model1 = model1.predict(x_test)
y_pred_model2 = model2.predict(x_test)
y_pred_model3 = model3.predict(x_test)
y_pred_model4 = model4.predict(x_test)
y_pred_model5 = model5.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accurecy of the model1 is : {0}%".format(accuracy_score(y_test,y_pred_model1)*100))
print("Accurecy of the model2 is : {0}%".format(accuracy_score(y_test,y_pred_model2)*100))
print("Accurecy of the model3 is : {0}%".format(accuracy_score(y_test,y_pred_model3)*100))
print("Accurecy of the model4 is : {0}%".format(accuracy_score(y_test,y_pred_model4)*100))
print("Accurecy of the model5 is : {0}%".format(accuracy_score(y_test,y_pred_model5)*100))
```
